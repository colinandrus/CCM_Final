```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mturk_df = pd.read_csv('Batch_3628097_batch_results (1).csv')
```

```python
mturk_df.columns
```

```python
mturk_df = mturk_df.drop(['HITId', 'HITTypeId', 'Title', 'Description', 'Keywords', 'Reward',
       'CreationTime', 'MaxAssignments', 'RequesterAnnotation',
       'AssignmentDurationInSeconds', 'AutoApprovalDelayInSeconds',
       'Expiration', 'NumberOfSimilarHITs', 'LifetimeInSeconds',
       'AssignmentId', 'WorkerId', 'AssignmentStatus', 'AcceptTime',
       'SubmitTime', 'AutoApprovalTime', 'ApprovalTime', 'RejectionTime',
       'RequesterFeedback', 'WorkTimeInSeconds', 'LifetimeApprovalRate',
       'Last30DaysApprovalRate', 'Last7DaysApprovalRate', 'Approve', 'Reject'], axis=1)
mturk_df
```

```python
mturk_df['Input.pair1_image1'][0].replace('https://s3.us-east-2.amazonaws.com/bird-bucket-ccm/', '')
```

```python
image_col_list = ['Input.pair1_image1',
       'Input.pair1_image2', 'Input.pair2_image1', 'Input.pair2_image2',
       'Input.pair3_image1', 'Input.pair3_image2', 'Input.pair4_image1',
       'Input.pair4_image2', 'Input.pair5_image1', 'Input.pair5_image2']
for col in image_col_list:
    mturk_df[col] = mturk_df[col].str.replace('https://s3.us-east-2.amazonaws.com/bird-bucket-ccm/', '')
mturk_df
```

```python
mturk_df['comparison1'] = mturk_df[['Input.pair1_image1','Input.pair1_image2']].apply(lambda x: '_'.join(x,), axis=1)
mturk_df['comparison2'] = mturk_df[['Input.pair2_image1','Input.pair2_image2']].apply(lambda x: '_'.join(x), axis=1)
mturk_df['comparison3'] = mturk_df[['Input.pair3_image1','Input.pair3_image2']].apply(lambda x: '_'.join(x), axis=1)
mturk_df['comparison4'] = mturk_df[['Input.pair4_image1','Input.pair4_image2']].apply(lambda x: '_'.join(x), axis=1)
mturk_df['comparison5'] = mturk_df[['Input.pair5_image1','Input.pair5_image2']].apply(lambda x: '_'.join(x), axis=1)
```

```python
mturk_df = mturk_df.drop(['Input.pair1_image1',
       'Input.pair1_image2', 'Input.pair2_image1', 'Input.pair2_image2',
       'Input.pair3_image1', 'Input.pair3_image2', 'Input.pair4_image1',
       'Input.pair4_image2', 'Input.pair5_image1', 'Input.pair5_image2'], axis=1)
```

```python
mturk_df.columns
```

```python
long_df = pd.DataFrame({'comparison':pd.concat([mturk_df['comparison1'],
                           mturk_df['comparison2'],
                           mturk_df['comparison3'],
                           mturk_df['comparison4'],
                           mturk_df['comparison5']], ignore_index=True) , 
                          'similarity': pd.concat([mturk_df['Answer.Similarity 1'],
                              mturk_df['Answer.Similarity 2'],
                              mturk_df['Answer.Similarity 3'],
                              mturk_df['Answer.Similarity 4'],
                              mturk_df['Answer.Similarity 5']], ignore_index=True)
                          })
```

```python
long_df.groupby('comparison').mean()
```

```python
long_df[long_df['comparison'] == 'image_6.jpg_image_9.jpg'].hist()
plt.title('Images 6 and 9 Similarty Score Histogram')
plt.xlabel('Similarity')
plt.ylabel('Count')
plt.grid(False)
plt.xlim(0,10)
plt.show()
```
